```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/nickel/question 2/StudentsPerformance.csv")
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Check for duplicates
print(df.duplicated().sum())
```

```python
# Remove whitespace from column names
df.columns = df.columns.str.strip()
```

```python
# Drop rows with null values
df.dropna(inplace=True)
```

```python
# Convert lunch values to lowercase and replace misspellings
df['lunch'] = df['lunch'].str.lower().replace('standar', 'standard')
```

```python
df.to_csv("C:/nickel/question 2/cleaned_student_performance.csv", index=False)
```

```python
df = pd.read_csv("C:/nickel/question 2/cleaned_student_performance.csv")
```

```python
# Visualization 1: Histogram of Math Scores
plt.hist(df['math score'], bins=20)
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Distribution of Math Scores')
plt.show()
```

```python
# Visualization 2: Scatterplot of Math and Reading Scores
plt.scatter(df['math score'], df['reading score'])
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.title('Relationship between Math and Reading Scores')
plt.show()
```

```python
# Visualization 3: Barplot of Test Preparations by Gender
sns.countplot(x='test preparation course', hue='gender', data=df)
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.title('Test Preparation by Gender')
plt.show()
```

```python
# Visualization 4: Boxplot of Math Scores by Ethnicity
sns.boxplot(x='race/ethnicity', y='math score', data=df)
plt.xlabel('Ethnicity')
plt.ylabel('Math Score')
plt.title('Math Scores by Ethnicity')
plt.show()
```

```python
# Visualization 5: Barplot of Parental Education by Lunch Type
sns.countplot(x='parental level of education', hue='lunch', data=df)
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.title('Parental Education by Lunch Type')
plt.show()
```
